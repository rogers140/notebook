```python
import sys
import os
sys.path.append(os.path.join(os.environ['GORDON_REPO_DIR'], 'utilities'))
from utilities2015 import *

from matplotlib.path import Path
%matplotlib inline
```

```python
# Accuracy with stride
stride = [2, 4, 8, 16, 32, 64];
train_acc_stride = [84.6, 83.9, 83.8, 79.1, 76.0, 63.0];
test_acc_stride = [61.4, 60.3, 60.6, 55.4, 52.2, 43.5];
ax1 = plt.subplot(111);
plt.xlabel("Stride between sample points");
plt.ylabel("Prediction accuracy (%)");
plt.plot(stride, train_acc_stride, label='training');
plt.plot(stride, test_acc_stride, label='testing');
plt.legend();
```

```python
# Accuracy with Level when stride=4, M = 200
L = [0, 1, 2, 3];
train_acc_L = [74.4, 79.9, 83.9, 87.4];
test_acc_L = [62.4, 38.4, 60.3, 59.8];
ax2 = plt.subplot(111);
plt.xlabel("Total number of levels (from 0)");
plt.ylabel("Prediction accuracy (%)");
plt.plot(L, train_acc_L, label='training');
plt.plot(L, test_acc_L, label='testing');
plt.legend();
```

```python
# Accuracy with M (feature size) when stride=8, L=2
M = [200, 300, 400]
train_acc_M = [83.8, 86.5, 88.3];
test_acc_M = [60.6, 61.1, 64.2];
ax3 = plt.subplot(111);
plt.xlabel("Number of features channels (M)");
plt.ylabel("Prediction accuracy (%)");
plt.plot(M, train_acc_M, label='training');
plt.plot(M, test_acc_M, label='testing');
plt.legend();
```
